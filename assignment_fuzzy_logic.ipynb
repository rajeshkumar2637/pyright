#   1. Let A(x) and B(x) be two fuzzy sets of speed limit on a highway, defined by the following membership functions:

```python
def membership_A(x):
    if x<30:
        return 1
    elif 30<=x<50:
        return ((x-30)/20)
    elif 50<=x<70:
        return ((x-50)/20)
    else:
        return 0


def membership_B(x):
    if x<60:
        return 0
    elif 60<=x<80:
        return ((x-60)/20)
    elif 80<=x<100:
        return ((x-80)/20)
    else:
        return 1
```

```python
listA=[]
for i in range(0, 30):
    listA.append(membership_A(i))

for i in range(30, 130):
    listA.append(0)


listB=[]
for i in range(0, 30):
    listB.append(0)

for i in range(30, 50):
    listB.append(membership_A(i))

for i in range(50, 130):
    listB.append(0)


listC=[]
for i in range(0, 50):
    listC.append(0)

for i in range(50, 70):
    listC.append(membership_A(i))

for i in range(70, 130):
    listC.append(0)


listD=[]
for i in range(0, 70):
    listD.append(0)

for i in range(70, 70):
    listD.append(membership_A(i))

for i in range(70, 130):
    listD.append(0)



listE=[]
for i in range(0, 60):
    listE.append(membership_B(i))

for i in range(60, 130):
    listE.append(0)



listF=[]
for i in range(0, 60):
    listF.append(0)

for i in range(60, 80):
    listF.append(membership_B(i))

for i in range(80, 130):
    listF.append(0)


listG=[]
for i in range(0, 80):
    listG.append(0)

for i in range(80, 100):
    listG.append(membership_B(i))

for i in range(100, 130):
    listG.append(0)


listH=[]
for i in range(0, 100):
    listH.append(0)

for i in range(100, 130):
    listH.append(membership_B(i))
```

#   2. Let C(x) and D(x) be two fuzzy sets of temperature (Celsius) in a city, defined by the following membership functions:

```python

# Define the membership functions for C (Cold) and D (Warm)
def mu_C(x):
    if x < -10:
        return 0
    elif -10 <= x < 0:
        return (x + 10) / 10
    elif 0 <= x <= 5:
        return 1
    else:
        return 0

def mu_D(x):
    if x < 25:
        return 0
    elif 25 <= x < 35:
        return (x - 25) / 10
    elif 35 <= x <= 40:
        return 1
    else:
        return 0
```

#   calculating Max-Min compostiion and their Graph

```python


# Define the universe of discourse
X_C = np.arange(-15, 10, 0.1)  # For fuzzy set C (Cold)
X_D = np.arange(20, 45, 0.1)   # For fuzzy set D (Warm)

# Calculate the fuzzy relation R (C x D)
R = np.zeros((len(X_C), len(X_D)))

for i, x_C in enumerate(X_C):
    for j, x_D in enumerate(X_D):
        R[i, j] = min(mu_C(x_C), mu_D(x_D))



# Redefine the arrays for the compositions
max_min_comp1 = np.zeros(len(X_C))



# Max-Min Composition
for i, x_C in enumerate(X_C):
    max_min_comp1[i] = np.max(np.minimum(mu_C(x_C), R[i, :]))


# Visualization
plt.figure(figsize=(8, 8))

# Plot the Max-Min Composition

plt.plot(X_C, max_min_comp1, label="Max-Min Composition (C  D)", color='black')
plt.title("Max-Min Composition(a, C  D)")

plt.xlabel("Temperature (C)")
plt.ylabel("Membership Degree")
plt.grid(True)
plt.legend()


plt.tight_layout()
plt.show()

```

#   calculating Max-product compostiion and their Graph

```python
# Redefine the arrays for the compositions
max_prod_comp = np.zeros(len(X_C))

# Max-Product Composition
for i, x_C in enumerate(X_C):
    max_prod_comp[i] = np.max(mu_C(x_C) * R[i, :])

# Plot the Max-Product Composition

plt.plot(X_C, max_prod_comp, label="Max-Product Composition (C  D)", color='red')
plt.title("Max-Product Composition(a, C  D)")
plt.xlabel("Temperature (C)")
plt.ylabel("Membership Degree")
plt.grid(True)
plt.legend()


plt.tight_layout()
plt.show()
```
